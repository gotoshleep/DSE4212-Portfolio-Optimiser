```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
training_data = pd.read_csv('training_data.csv')
covid_data = pd.read_csv('covid_stress_test_data.csv')
```

```python
# Combine (stack) them row-wise
merged_data = pd.concat([training_data, covid_data], ignore_index=True)

# Ensure 'Date' column is properly recognized as datetime
merged_data['Date'] = pd.to_datetime(merged_data['Date'])

# Sort chronologically
merged_data = merged_data.sort_values(by='Date').reset_index(drop=True)
```

```python
corr_matrix_3 = merged_data.corr(numeric_only=True)

# %% Plot correlation heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(
    corr_matrix_3,
    annot=False,         
    fmt=".2f",           # Format to 2 decimal places
    cmap='coolwarm',     
    center=0,            
    linewidths=0.5,      
    square=True          
)

plt.title('Correlation Heatmap', fontsize=14, pad=12)
plt.tight_layout()
plt.show()

corr_matrix_3
```
